```python
# Expandindo os dados e verificando os nomes reais das colunas
dados_expandidos = pd.json_normalize(dados.to_dict(orient='records'), sep='_')

# Vamos listar as colunas para conferir os nomes exatos
print("Colunas disponíveis após a expansão:")
print(dados_expandidos.columns.tolist())

# Tentativa de visualização inicial para validar os nomes
dados_expandidos.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Preparando os dados para o gráfico
# Agrupamos por Churn e Tenure para ver a soma do faturamento total
plot_data = dados_expandidos.groupby(['Churn', 'customer_tenure'])['account_Charges_Total'].sum().unstack(level=0)

# Criando a visualização
plt.figure(figsize=(12, 6))
sns.set_style("whitegrid")

plt.plot(plot_data.index, plot_data['No'], label='Retidos (Fiel)', marker='o', color='#1f77b4', linewidth=2, alpha=0.8)
plt.plot(plot_data.index, plot_data['Yes'], label='Churn (Cancelado)', marker='x', color='#ff7f0e', linestyle='--', linewidth=2, alpha=0.8)

# Personalização do gráfico
plt.title('Impacto Financeiro: Faturamento Acumulado vs. Tempo de Contrato', fontsize=15, pad=20)
plt.xlabel('Meses de Contrato (Tenure)', fontsize=12)
plt.ylabel('Soma de Faturamento Total ($)', fontsize=12)
plt.legend(title='Status do Cliente', frameon=True)
plt.grid(True, linestyle='--', alpha=0.5)

# Adicionando uma anotação sobre a estagnação do churn
plt.annotate('Estagnação de faturamento por Churn',
             xy=(20, plot_data['Yes'].iloc[20]),
             xytext=(35, 150000),
             arrowprops=dict(facecolor='black', shrink=0.05, width=1),
             fontsize=10)

plt.tight_layout()
plt.show()
```

#📊 Carga e análise

```python
# Convertendo a coluna de cobranças para numérico antes de somar
# O parâmetro errors='coerce' transformará eventuais espaços vazios em NaN
dados_expandidos['account_Charges_Total'] = pd.to_numeric(dados_expandidos['account_Charges_Total'], errors='coerce')

# Realizando o agrupamento numérico
analise_churn_tenure = dados_expandidos.groupby(['Churn', 'customer_tenure'])[['account_Charges_Total']].sum()
analise_churn_tenure.head(10)
```
